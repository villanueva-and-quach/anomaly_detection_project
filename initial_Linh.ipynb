# 2. Is there a cohort that referred to a lesson significantly more than other cohorts seemed to gloss over?

```python
#Using this function to count most visited and least visited
from collections import Counter
 
def most_frequent(List):
    occurence_count = Counter(List)
    return occurence_count.most_common(1)[0][0]
```

```python
#Creating loops for program id and cohort to count most frequent visit and least frequent visit

cohort_type_list= list(df['program_id'].dropna().unique())
cohort_list= list(df['cohort_id'].dropna().unique())

for type_of_cohort in cohort_type_list:
    print("################################################")
    print(f"{type_of_cohort}")
    print("################################################")
    most_visited=[]
    least_visited=[]
    for cohort in cohort_list:
        #filter 1: type of cohort (data science vs web dev)
        isolated_df= df[df['program_id'] == type_of_cohort]
        #filter 2: cohort id (ada vs florence)
        isolated_df= isolated_df[isolated_df['cohort_id'] == cohort]

        #Check if new dataframe is empty
        if isolated_df.shape[0] > 0:
            num_of_records= isolated_df.shape[0]
            #Percentage of times page visited
            percent_visited= (isolated_df['endpoint'].value_counts())/num_of_records
            #append most_visited to most_visited list
            most_visited.append(pd.DataFrame(percent_visited).reset_index().iloc[0]['index'])
            least_visited.append(pd.DataFrame(percent_visited).reset_index().iloc[-1]['index'])
            
    print(most_frequent(most_visited))
    print(most_frequent(least_visited))
```

When comparing cohorts across programs you can see the many of the cohorts reviewed javascript-I, data science fundamentals, html-css content, and modern data scientist fundamentals.
