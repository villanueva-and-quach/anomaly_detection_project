```python
#reset index to make list
new= cohort_pv_df.reset_index()
```

```python
#Making cohort list
cohort_list =list(new.cohort_id.value_counts().index)
```

```python
#Loop to to see most pages viewed by cohort

most_pageviews_df= pd.DataFrame(columns={'cohort_id'})

for cohort in cohort_list:
    var= cohort_pv_df.loc[cohort]
    y = var.nlargest(1, 'endpoint_count')
#     most_pageviews_df['cohort_id']= cohort
    most_pageviews_df= pd.concat([most_pageviews_df, y], ignore_index = False)
```

```python
#Looking at values after for loop
most_pageviews_df
```

```python
#set cohort id column to cohort list
most_pageviews_df['cohort_id']= cohort_list
```

```python
#sorting values, resetting index, set_index, and renaming columns
most_pageviews_df= most_pageviews_df.sort_values(by='cohort_id').reset_index().set_index('cohort_id').rename(columns={'index':'endpoint'})
```

```python
#Looking at values
most_pageviews_df
```

```python
most_pageviews_df.groupby('endpoint').count().sort_values(by='endpoint_count', ascending= False)
```

```python
#Loop to to see least pages viewed by cohort

least_pageviews_df= pd.DataFrame(columns={'cohort_id'})

for cohort in cohort_list:
    var= cohort_pv_df.loc[cohort]
    y2 = var.nsmallest(1, 'endpoint_count')
#     most_pageviews_df['cohort_id']= cohort
    least_pageviews_df= pd.concat([least_pageviews_df, y2], ignore_index = False)
```
